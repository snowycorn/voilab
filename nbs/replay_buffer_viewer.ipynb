```python
import ipywidgets as w
from voilab.applications import replay_buffer_viewer
```

# Replay Buffer Viewer
> A lightweight visualization tool for exploring datasets generated by the Universal Manipulation Interface (UMI). It supports the latest replay buffer structure, allowing users to easily inspect time-series data, including RGB camera streams, robot end-effector positions and orientations, gripper widths, and demonstration start/end poses. Ideal for debugging, validation, and quick data analysis in robot learning pipelines.

```python
data_path_widget = w.Text(value="../video/dataset.zarr.zip", description="Zarr path:", layout=w.Layout(width="400px"))
run_button = w.Button(description="Load", button_style="success")
output = w.Output()


form = w.VBox([data_path_widget, run_button, output])
display(form)


def load_and_show(_):
    with output:
        replay_buffer_viewer.show(data_path_widget.value)


run_button.on_click(load_and_show)
```
